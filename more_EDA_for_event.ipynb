### Step 3: Load the Data Section



```python
import pandas as pd

file_path = '/content/drive/My Drive/MATCH-RESULTS.csv'
df = pd.read_csv(file_path)
```

```python
# Number of home goals of real madrid per referee

#Filter only for Real Madrid
target_value = 'Real Madrid'
real_df = df[df['Home'] == target_value]

grouped_df = real_df.groupby('Referee')['HomeGoals'].sum()
grouped_df = pd.DataFrame(grouped_df)
sns.barplot(x='Referee', y='HomeGoals', data=grouped_df)
plt.xticks(rotation=90)
plt.title('Home Goals of Real Madrid per Referee')
plt.xlabel('Referee')
plt.ylabel('Home Goals')
```

```python
# Number of away goals of real madrid per referee

#Filter only for Real Madrid
target_value = 'Real Madrid'
real_df2 = df[df['Away'] == target_value]


grouped2_df = real_df2.groupby('Referee')['AwayGoals'].sum()
grouped2_df = pd.DataFrame(grouped2_df)
sns.barplot(x='Referee', y='AwayGoals', data=grouped2_df)
plt.xticks(rotation=90)
plt.title('Away Goals of Real Madrid per Referee')
plt.xlabel('Referee')
plt.ylabel('Home Goals')
plt.show()
```

```python
# Number of home goals of barcelona per referee

#Filter only for Barcelona
target_value = 'Barcelona'
barca_df = df[df['Home'] == target_value]

grouped3_df = barca_df.groupby('Referee')['HomeGoals'].sum()
grouped3_df = pd.DataFrame(grouped3_df)
sns.barplot(x='Referee', y='HomeGoals', data=grouped3_df)
plt.xticks(rotation=90)
plt.title('Home Goals of Barcelona per Referee')
plt.xlabel('Referee')
plt.ylabel('Home Goals')
plt.show()
```

```python
# Number of home goals of real madrid per referee

#Filter only for Barcelona
target_value = 'Barcelona'
barca_df2 = df[df['Away'] == target_value]

grouped4_df = barca_df2.groupby('Referee')['AwayGoals'].sum()
grouped4_df = pd.DataFrame(grouped4_df)

sns.barplot(x='Referee', y='AwayGoals', data= grouped4_df)
plt.xticks(rotation=90)
plt.title('Away Goals of Barcelona per Referee')
plt.xlabel('Referee')
plt.ylabel('Away Goals')
plt.show()
```
